```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Initialize variables to store results
k_values = list(range(1, 21))
avg_accuracies = []

# Repeat the KNN classification for each k value
for k in k_values:
    accuracies = []
    
    # Repeat the classification at least five times
    for _ in range(5):
        # Split the dataset into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.66, random_state=42)
        
        # Create a KNN classifier with the current k value
        knn = KNeighborsClassifier(n_neighbors=k)
        
        # Fit the classifier to the training data
        knn.fit(X_train, y_train)
        
        # Calculate accuracy on the test data
        accuracy = knn.score(X_test, y_test)
        accuracies.append(accuracy)
    
    # Calculate the average accuracy for the current k value
    avg_accuracy = np.mean(accuracies)
    avg_accuracies.append(avg_accuracy)

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(k_values, avg_accuracies, marker='o')
plt.title('KNN Accuracy vs. K Value')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Average Accuracy')
plt.grid(True)
plt.xticks(k_values)
plt.show()

# Find the best K value with the highest average accuracy
best_k = k_values[np.argmax(avg_accuracies)]
print(f"The best K value is {best_k} with an average accuracy of {max(avg_accuracies)}")

```
